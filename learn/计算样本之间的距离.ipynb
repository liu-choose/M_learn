# 欧拉距离

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[5,1], 
            [1,4]])    
plt.plot(X[:,0],X[:,1],'r-o')
plt.show()
```

```python
plt.plot(X[:,0],X[:,1],'r-o')
plt.annotate('a',xy=X[0])
plt.annotate('b',xy=X[1])
plt.plot([1,5],[1,1],'g--')
plt.plot([1,1],[1,4],'g--')
plt.show()
```

# 明可夫斯基距离

```python
#默认取欧拉距离
def distance(a,b,p=2):
    return (np.sum(np.abs(a - b )**p))**(1/p)
```

```python
distance(X[0],X[1])
```

```python
distance(X[0],X[1],p=1)
```

```python
X = np.array([[5,1,11,6],
           [1,4,43,99]
          ])
```

```python
distance(X[0],X[1],p=1)
```
