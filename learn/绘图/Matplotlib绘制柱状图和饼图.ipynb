```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = ['Q1','Q2','Q3','Q4']
y = [10,30,20,60]
plt.bar(x,y,color='g',width=0.3)
plt.grid(True)
plt.show()
```

```python
x = ['Q1','Q2','Q3','Q4']
y = [10,30,20,60]
rect = plt.bar(x,y,color='g',width=0.3)
#print(len(rect))
for ind,item in enumerate(rect):
    _x = item.get_x() + 0.1
    _y = item.get_height() + 1
    plt.text(_x,_y,y[ind])
plt.ylim(0,70)
plt.grid(True)
plt.show()
```

```python
y = [10,30,20,60]
plt.pie(y,labels=x)
plt.show()
```

```python
y = [10,30,20,60]
plt.axes(aspect=1)
plt.pie(y,labels=x,autopct="%.2f%%",explode=[0.2,0,0.1,0.1],shadow=True)

plt.show()
```
