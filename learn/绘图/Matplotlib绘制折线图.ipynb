```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
X = np.arange(10)
X
```

```python
np.random.seed(1)
Y = np.random.random(size=10)
Y
```

```python
plt.plot(X,Y,color='r',marker='*',linestyle='--')
plt.show()
```

```python
plt.plot(X,Y,color='r',marker='*',linestyle='-.')
plt.show()
```

```python
plt.plot(X,Y,'g-.o')
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y,label='sin(x)')
plt.plot(x,y2,label='cos(x)')
plt.legend()
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title('My Demo')
plt.axis([5,])
plt.show()
```
